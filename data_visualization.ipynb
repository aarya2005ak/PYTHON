```python
import pandas
import matplotlib.pyplot as plt
import seaborn
```

```python
data={
    'Name':['Tom','nick','krish','jack','Ak'],
    'Percentage':[90,75,80,62,100]
}
```

```python
df=pandas.DataFrame(data)
print(df)
```

```python
plt.plot(df['Name'],df['Percentage'])
plt.title('Student Percentage')
plt.xlabel('Name')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()
```

```python
sizes=df['Percentage']
labels=df['Name']
plt.pie(sizes,labels=labels)
plt.title('Student Percentage')
plt.show()
```

```python
seaborn.barplot(x='Name',y='Percentage',data=df)
plt.title('Student Percentage')
plt.show()
```

```python
seaborn.lineplot(x='Name',y='Percentage',data=df)
plt.title('Student Percentage')
plt.show()
```
